# Network Structure

```python
class CNA(nn.Layer):
    #
    def __init__(self, in_channels, out_channels, kernel_size, stride, padding, conv, norm, act, bias):
        super().__init__()
        if conv == 'conv':
            self.layers = nn.LayerList([nn.Conv2D(in_channels, out_channels, kernel_size, stride, padding, weight_attr=nn.initializer.Normal(0., 0.02), bias_attr=bias)])
        if conv == 'tconv':
            self.layers = nn.LayerList([nn.Conv2DTranspose(in_channels, out_channels, kernel_size, stride, padding, weight_attr=nn.initializer.Normal(0., 0.02), bias_attr=bias)])
        if norm == 'bn':
            self.layers.append(nn.BatchNorm2D(out_channels))
        if act == 'relu':
            self.layers.append(nn.ReLU())
        if act == 'lrelu':
            self.layers.append(nn.LeakyReLU(0.2))
    #
    def forward(self, x):
        for layer in self.layers:
            x = layer(x)
        return x
```

## Generator

```python
class Generator(nn.Layer):
    #
    def __init__(self, dim_latent=dim_latent, channel=channel_g):
        super().__init__()
        self.layers = nn.Sequential(
            CNA(dim_latent, channel*8, 4, 1, 0, 'tconv', 'bn', 'relu', False),
            CNA(channel*8, channel*4, 4, 2, 1, 'tconv', 'bn', 'relu', False),
            CNA(channel*4, channel*2, 4, 2, 1, 'tconv', 'bn', 'relu', False),
            CNA(channel*2, channel, 4, 2, 1, 'tconv', 'bn', 'relu', False),
            CNA(channel, 3, 4, 2, 1, 'tconv', False, False, False),
            nn.Tanh()
        )
    #
    def forward(self, x):
        return self.layers(x)
```

## Discriminator

```python
class Discriminator(nn.Layer):
    #
    def __init__(self, channel=channel_d):
        super().__init__()
        self.layers = nn.Sequential(
            CNA(3, channel, 4, 2, 1, 'conv', False, 'lrelu', True),
            CNA(channel, channel*2, 4, 2, 1, 'conv', False, 'lrelu', True),
            CNA(channel*2, channel*4, 4, 2, 1, 'conv', False, 'lrelu', True),
            CNA(channel*4, channel*8, 4, 2, 1, 'conv', False, 'lrelu', True),
            CNA(channel*8, 1, 4, 1, 0, 'conv', False, False, False)
        )
    #
    def forward(self, x):
        return self.layers(x).reshape([-1,1])
```

# Train

```python
def gradient_penalty(y, x):
    """Compute gradient penalty: (L2_norm(dy/dx) - 1)**2."""
    # weight = P.ones(y.shape)
    dydx = P.grad(y, x, retain_graph=True, create_graph=True)[0]
    dydx = dydx.reshape([dydx.shape[0], -1])
    dydx_l2norm = P.sqrt(P.sum(dydx**2, axis=1))
    return P.mean((dydx_l2norm-1)**2)
```

```python
def trainer_d(net_d, optimizer_d, net_g, image_real):
    #
    src_real = net_d(image_real)
    image_fake = net_g(P.randn([image_real.shape[0], dim_latent, 1, 1]))
    src_fake = net_d(image_fake.detach())
    # gp
    Î± = P.rand([image_real.shape[0], 1, 1, 1])
    image_fusion = Î±*image_real + (1-Î±)*image_fake
    src_fusion = net_d(image_fusion)
    #
    loss = P.mean(src_fake) - P.mean(src_real) + Î»_gp*gradient_penalty(src_fusion, image_fusion)
    #
    optimizer_d.clear_grad()
    loss.backward()
    optimizer_d.step()
```

```python
def trainer_g(net_g, optimizer_g, net_d, image_real):
    #
    image_fake = net_g(P.randn([image_real.shape[0], dim_latent, 1, 1]))
    src_fake = net_d(image_fake)
    #
    loss = -P.mean(src_fake)
    #
    optimizer_g.clear_grad()
    loss.backward()
    optimizer_g.step()
```

```python
def train(batch_size, epochs, load_model):
    #
    net_d = Discriminator()
    net_d.train()
    net_g = Generator()
    net_g.train()
    optimizer_d = P.optimizer.Adam(parameters=net_d.parameters(), learning_rate=learning_rate, beta1=0.5)
    optimizer_g = P.optimizer.Adam(parameters=net_g.parameters(), learning_rate=2*learning_rate, beta1=0.5)
    if load_model:
        net_d.set_state_dict(P.load(path_work + 'net_d.pdparams'))
        net_g.set_state_dict(P.load(path_work + 'net_g.pdparams'))
    #
    iteration = 0
    for epoch in range(epochs):
        dataloader = DataLoader(Dataset_GAN(), batch_size=batch_size, shuffle=True)
        for image in dataloader:
            trainer_d(net_d, optimizer_d, net_g, image)
            iteration += 1
            if iteration % step_g == 0:
                trainer_g(net_g, optimizer_g, net_d, image)
    #
            if iteration % 1000 == 0:
                print('Epoch: ', epoch, ', Iteration: ', iteration)
                P.save(net_d.state_dict(), path_work + 'net_d.pdparams')
                P.save(net_g.state_dict(), path_work + 'net_g.pdparams')
                show(net_g)
                net_g.train()
```

```python
train(batch_size=batch_size, epochs=5, load_model=False)
```
